```python
import pandas as pd
from pyabsa import ATEPCCheckpointManager

# Loading Dataset into a DataFrame
import pandas as pd


csv_file_path = '/content/processed_data.csv'


df = pd.read_csv(csv_file_path)



# Initialize the aspect extractor
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(
    checkpoint='english',
    auto_device=True
)

def extract_aspects(text):
    # Perform aspect extraction
    result = aspect_extractor.extract_aspect(
        inference_source=[text],
        pred_sentiment=True
    )
    # Extract and return the aspects
    return result[0]['aspect']

# Apply aspect extraction to each row in the DataFrame
df['aspects'] = df['text'].apply(extract_aspects)

# Display the DataFrame with the extracted aspects
print(df)
```

```python
# Specify the path where you want to save the CSV file
csv_file_path = 'withaspect.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)
```

The code snippet provided uses the PyABSA library's ATEPC (Aspect Term Extraction with Aspect and Sentiment Classification) model for aspect extraction, which is built on BERT (Bidirectional Encoder Representations from Transformers). PyABSA's ATEPC models leverage BERT or similar transformer-based architectures for high-performance aspect extraction and sentiment analysis.

BERT is indeed working in the background when we use the aspect_extractor from PyABSA. The ATEPCCheckpointManager initializes the model, which uses BERT's architecture to perform the aspect extraction.
